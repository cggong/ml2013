```python
# function for removing pairs of elements from numpy arrays that are Nan
def delete_nan(list_1, list_2):
    import numpy as np
    from numpy import float32
    temp_list = np.array([list_1, list_2], dtype=float32).transpose()
    return temp_list[~np.isnan(temp_list).any(1)].transpose()

# function of Gaussian Kernel
def gaussian_kernel(matrix_x):
    import numpy as np
    return np.exp( - matrix_x ** 2 / 2.0 )

# function of Epanechnikov kernel, here I put 'import numpy' twice because it is required when we try to do parallelization and deal with closures
def epanech_kernel(matrix_x):
    import numpy as np
    matrix_without_indicator = 1 - matrix_x ** 2
    # set negative entries to zero
    matrix_with_indicator = matrix_without_indicator.clip(min=0)
    return matrix_with_indicator

# function of Boxcar kernel
def boxcar_kernel(matrix_x):
    import numpy as np
    matrix_with_indicator = matrix_x
    
    start = datetime.datetime.now()
    matrix_with_indicator[matrix_with_indicator <= 1] = 1
    print datetime.datetime.now() - start
    
    start = datetime.datetime.now()
    matrix_with_indicator[matrix_with_indicator > 1] = 0
    print datetime.datetime.now() - start
    
    return matrix_with_indicator    

# function of Tricube kernel
def tricube_kernel(matrix_x):
    import numpy as np
    matrix_without_indicator = 1 - np.abs(matrix_x ** 3)
    # set negative entries to zero
    matrix_with_indicator = matrix_without_indicator.clip(min=0)
    return matrix_with_indicator

'''
dview['gaussian_kernel'] = gaussian_kernel
dview['epanech_kernel'] = epanech_kernel
dview['boxcar_kernel'] = boxcar_kernel
dview['tricube_kernel'] = tricube_kernel
'''
```

```python
#clear the cache of any inconsistencies; important for handling crashes
DAL.cleancache()
lightcurves = DAL.create('lightcurves')
s = lightcurves.subsets()

dat = []
# process data for 3 types of light curves, we use the examples given in the problem set
# pick up 'conf' curve from subset 100 (corresponding to '189.part.zip'), whose id is 914193
# pick up 'fp' curve from subset 299 (corresponding to '64.part.zip'), whose id is 6258272
# pick up 'ep' curve from subset 148 (corresponding to '231.part.zip'), whose id is 485585
for m in ['189.part.zip', '64.part.zip', '231.part.zip']:
    for i in lightcurves.iter(m):
        name = i['id']
        if (name in [914193, 6258272, 485585]):
            lc = i['data']
            time = lc[:len(lc)/2]  #modified - first half time
            flux = lc[len(lc)/2:]  #modified - first half flux
            # delete both time and flux if NAN appears in either of them
            temp_time_flux = delete_nan(time, flux)
            # output name, time, flux
            dat.append( (name, temp_time_flux[0], temp_time_flux[1]) )

'''
dview['dat'] = dat
'''
```

```python
# function of forming matrix L

def matrix_L(kernel, x, bandwidth):
    import numpy as np
    from numpy import float32
    
    #start = datetime.datetime.now()
    diff_matrix = np.zeros(len(x)**2, dtype = float32).reshape((len(x), len(x))) # initialization
    for i in xrange(len(x)):
        diff_matrix[i] = x - x[i]
    diff_matrix = diff_matrix / float(bandwidth)
    #print datetime.datetime.now() - start
    
    #start = datetime.datetime.now()
    if (kernel == 'Gaussian'):
        L_matrix  = gaussian_kernel(diff_matrix)
    if (kernel == 'Epanech'):
        L_matrix = epanech_kernel(diff_matrix)
    if (kernel == 'Boxcar'):
        L_matrix = boxcar_kernel(diff_matrix)
    if (kernel == 'Tricube'):
        L_matrix = tricube_kernel(diff_matrix)
    #print datetime.datetime.now() - start
    
    # list containing the sum over each row
    #start = datetime.datetime.now()
    sum_row = np.dot(L_matrix, np.ones(len(x)))
    #print datetime.datetime.now() - start
    
    #start = datetime.datetime.now()
    L_matrix = np.dot(np.diag( 1.0 / sum_row ) , L_matrix)
    #print datetime.datetime.now() - start
    
    return L_matrix
'''
dview['matrix_L'] = matrix_L
'''
```

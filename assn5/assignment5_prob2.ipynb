```python
#verify cluster connection
from IPython.parallel import Client
rc = Client()
print rc.ids
dview = rc[:]
```

```python
import numpy
import DAL

# We gonna work with 2000 images of laptop and 2000 images of socks
tinyimages = DAL.create('tinyimages')
laptop_ids = tinyimages.search('laptop', 2000)
sock_ids = tinyimages.search('sock', 2000)
ids = laptop_ids + sock_ids

def gist(index):
    import DAL
    import scipy
    import leargist
    tinyimages = DAL.create('tinyimages')
    img = scipy.misc.toimage(tinyimages.byid(index).reshape(32,32,3, order="F").copy())
    return leargist.color_gist(img)

# calculate our gists
gists = dview.map_sync(gist, ids)
dview['gists'] = gists


# Standardize the vectors to have mean zero and variance one
#for i in range(4000):
#    mean = numpy.mean(gists[i])
#    std = numpy.std(gists[i])
#    # necessary since ids[281] for sock/laptop is all black and has gist 0,...,0. 
#    # Avoid dividing by std = 0.
#    if std==0:
#        gists[i] -= mean
#    else:
#        gists[i] = (gists[i]-mean)/std

# 'gists' is a list of 4000 vectors with zero mean and unit veriance
```

```python
# euc_dist is a matrix including the Euclidean distance between pair of images
euc_dist = numpy.zeros(16000000).reshape((4000, 4000))

for i in range(4000):
    for j in range(i+1,4000):
        dist = numpy.linalg.norm(gists[i]-gists[j])
        euc_dist[i,j] = dist
        euc_dist[j,i] = dist
```

```python
# Initialize centers.
k = 10
centers = range(k)
centers[0] = gists[1]
centers[1] = gists[301]
centers[2] = gists[601]
centers[3] = gists[901]
centers[4] = gists[320]
centers[5] = gists[14]
centers[6] = gists[321]
centers[7] = gists[621]
centers[8] = gists[921]
centers[9] = gists[323]
```

```python
# returns id of closest center
def closest_center(gist):
    import numpy  
    
    dist_to_cents = []
    for center in centers:
        dist = numpy.linalg.norm(gist-center)
        dist_to_cents.append(dist)
            
    min_dist = 100
    index = -1
    
    for i in range(len(dist_to_cents)):
        cur_dist = dist_to_cents[i]
        if cur_dist < min_dist:
            min_dist = cur_dist
            index = i
        else:
            pass
       
    return index
```

```python
not_converge = True

while (not_converge):
    dview['centers']=centers
    result = dview.map_sync(closest_center,gists)
    
    # update centers to means of clusters
    # track amount changed
    change = 0
    newcenters = range(len(centers))
    for i in range(len(centers)):
        gists_in_cluster = numpy.array([gists[j] for j in range(len(gists)) if result[j] == i])
        newcenters[i] = numpy.mean(gists_in_cluster, axis = 0)
        change += numpy.linalg.norm(newcenters[i]-centers[i])
    
    centers=newcenters
    print change
    if change < 0.01:
        not_converge = False
    # does not converge monotonically!
```

It seems like the model is moderately sensitive to the initial choice of centers. This run with k=10 was carried out with the same settings as the previous k=10 run, but with the last center altered. Several of the clusters seem unchanged. That is, their images have the same distinguishing features as in the earlier run. For instance, Cluster 2's sample images are still mostly L shaped figures set on a white background. Cluster 6 continues to be have mirror image L figures also on a white background. Clusters 4,7 and 8 continue to be multicolored, texture images. These observations suggest that changing a single center's initial setting is not enough to perturb the eventual locations of the other centers. 

However, changing a center's initial location seems to lead to a different cluster. Cluster 9 in the previous run shows a collection of socks. Or more generally, its images are characterised by having few pastel colors. However, with the initial location altered, Cluster 9 now shows a collection of laptops. The images now consist of more colors consisting whites, greys, blacks, and blues.

```python
from scipy.stats import norm
import numpy as np
```

```python
# calculate the empirical probabilities that the max's over T trials are more than the threshold 
def emp_prob(n,r,maxlist):
    threshold = np.sqrt(2*r*log(n))
    return np.mean(map(lambda x: x>= threshold, maxlist))
```

```python
T = 1000
K = range(5,16)
N = map(lambda x: 2**x, K)

nstore = []
xcoord = np.arange(75)/25.0
nappend = nstore.append
for n in N:
    # generate huge matrix with normal rvs for all trials 
    trials = norm.rvs(size=n*T).reshape((T,n))
            
    maxlist = map(max,abs(trials))
    pnrstore = []
    pnrappend = pnrstore.append
    for r in xcoord:
        pnrappend(emp_prob(n,r,maxlist))
    nappend(pnrstore)
```

```python
# plot all curves on the same graph
import matplotlib.pyplot as plt

plt.figure(figsize=(20,8))
for pnrstore in nstore:
    plt.plot(xcoord,pnrstore)

leg = map(lambda x: "n = 2**"+str(x),K)
plt.legend(leg, loc='upper right')
plt.title("Empirical probabilities against r for different no. of samples n")
plt.xlabel("r")
plt.ylabel("p_n(r)")
plt.show()
```
